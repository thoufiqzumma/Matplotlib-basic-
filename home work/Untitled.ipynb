```python
import matplotlib.pyplot as pt #importing matplotlib pt is an alias of pyplot
import pandas as pd #importing pandas , pd is an elias of pandas
#If we want to read something from dataset or csv file we use read_csv() method.
data= pd.read_csv("COVID-19-Bangladesh.csv")
data= data.head(40) # head() method is used to select some elements of dataset
#here i am plotting rollno vs cgpa so x-axis will be rollno and y-axis will be cgpa
#color of scatter is blue and we also include label as scatter.
#scatter() method is used to plot the scatter
pt.scatter(data["total_confirmed"],data["total_recovered"],color="blue",label="scatter")
pt.xlabel("total_confirmed",color="green") #xlabel() defines the label of x-axis
pt.ylabel("total_recovered",color="blue") #ylabel() defines the label of x-axis
pt.title("total_confirmed vs total_recovered",color="green") #title() is used to give title of this scatter plot
pt.show()
```

# Bar plot

```python
import matplotlib.pyplot as pt
import pandas as pd
data = pd.read_csv("COVID-19-Bangladesh.csv")
data= data.head(30)
#bar() method is used to plot a bar graph
#Here i am taking a list of colors to showing graph attractive
pt.bar(data["total_confirmed"],data["total_deaths"],color=["green","blue","pink","red"])
pt.xlabel("total_confirmed",color="green")
pt.ylabel("total_deaths",color="blue")
pt.title("total_deaths vs total_confirmed",color="green")
pt.show()
```

# pie chart

```python
import matplotlib.pyplot as pt
import pandas as pd
data = pd.read_csv("COVID-19-Bangladesh.csv")
data = data.head(30)
x=len(data[data.total_deaths>=1]) #students having cgpa over 9 point
x1=len(data[(data.total_deaths>=4) & (data.total_deaths<8)]) #students got over 8 point but less than 9 point
x2=len(data[data.total_deaths<5])
x3=len(data[data.total_deaths>=8])
#students having cgpa less than 8 point
pt.axis('equal') #for making pie chart circular,that makes major axis and minor axis equal
#Here we need a list of values that are simply x,x1 and x2
#colors specify a list of colors in pie chart
#In order to specify labels we use labels attribute
pt.pie([x,x1,x2],colors=['yellow','red','blue'],labels=['per day','per day','others'])
pt.legend(title='Description') # to shown the labels as legends
pt.show()
```
